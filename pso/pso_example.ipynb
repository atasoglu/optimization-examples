```python
import numpy as np
import matplotlib.pyplot as plt
from pso import Swarm
```

```python
# parameters
size = 100
x_bounds = y_bounds = [0, 100]
target = np.array([50, 50])
iteration = 20
w, c1, c2 = 0.5, 1, 2
velocity_factor = 0.4
random_state = 42
```

```python
# define fitness function
def fit_func(pos): # euclidean distance
    sub = pos - target
    return np.sqrt(sub[0]**2 + sub[1]**2)
```

```python
# create swarm
swarm = Swarm(size, x_bounds, y_bounds, w, c1, c2, velocity_factor, random_state)

# get initial xy values for evaluation
init_xv, init_yv = swarm.get_pos_values()

# run pso algorithm
for _ in range(iteration):
    swarm.calc_fitness(fit_func)
    swarm.step()

# get final xy values
xv, yv = swarm.get_pos_values()
```

```python
# plotting
_, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].scatter(init_xv, init_yv, 25, "g", "o", label="initial position")
axs[1].scatter(xv, yv, 25, "b", "o", label="final position")
axs[1].axvline(target[0], c="r", zorder=0)
axs[1].axhline(target[1], c="r", zorder=0)
axs[0].set_title("initial positions")
axs[1].set_title("final positions")
plt.tight_layout()
# plt.show()
```
